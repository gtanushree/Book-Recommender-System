```python
# Constructing Popularity Based Recommender System

ratings_with_name = ratingsData.merge(bookData, on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'NumRating'}, inplace=True)
num_rating
```

```python
avg_rating = ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'AvgRating'}, inplace=True)
avg_rating
```

```python
Popular = num_rating.merge(avg_rating, on='Book-Title')
Popular
```

```python
Popular = Popular[Popular['NumRating']>=250].sort_values('AvgRating', ascending=False)
Popular = Popular.merge(bookData, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','NumRating','AvgRating']]
Popular
```

```python
pickle.dump(Popular,open('Popular.pkl','wb'))
```
