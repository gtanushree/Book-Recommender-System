```python
# Constructing Collaborative Filtering Based Recommender System

ratings_with_name = ratingsData.merge(bookData, on='ISBN')
ratings_with_name
```

```python
a = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
active_users = a[a].index #boolean indexing
active_users
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
```

```python
b = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] > 50
famousBooks = b[b].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famousBooks)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
pt
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = bookData[bookData['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
